# Negative Image  
## Method 1.

```python
import numpy as np
```

```python
img_neg = np.zeros(img.shape, dtype=np.uint8)
```

```python
img_neg.dtype
```

```python
img.shape[0]
```

```python
for i in range(img.shape[0]): # Rows ## Iterates through each pixel on the row
    for j in range(img.shape[1]): # Columns ## Iterates through each pixel on the column
        img_neg[i, j] = 255 - img[i, j] # Changes the value to its negative
```

```python
# Prints the image
plt.imshow(img_neg, cmap="gray")
plt.axis("off")
plt.show()
```

```python
plt.imshow(np.hstack((img, img_neg)), "gray")
plt.title("Comparison of the transformation")
plt.show()
```

# Negative Transformation  
  
## Method 2 - BITWISE NOT

```python
img2 = cv.imread(r"..\Imagenes\squirrel_cls.jpg", 0)
```

```python
type(img2)
```

```python
img2.shape
```

```python
plt.imshow(img2, "gray") # We can skip cmap= because is the second parameter that imshow recieves
plt.title("Grayscale image") # Adds a title to the image
plt.axis("off") # Removes the axis of the image
plt.show()
```

```python
img2_neg = np.bitwise_not(img2)

# Inverts the value of each bit on each location of every pixel, resulting in a negative image as a result.
```

```python
plt.imshow(img2_neg, "gray")
plt.title("Negative image")
plt.axis("off")
plt.show()
```

```python
plt.imshow(np.hstack((img2, img2_neg)), "gray")
plt.title("Comparison of both images")
plt.show()
```
