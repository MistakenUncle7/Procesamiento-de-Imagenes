# **Gamma Transformation**

```python
import math
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread(r"..\Images\squirrel_cls.jpg", 0)
```

```python
plt.imshow(img, cmap="gray")
plt.title("Squirrle on grayscale")
plt.axis("off")
plt.show()
```

```python
img_gamma = img.copy()
```

```python
rows, columns = img_gamma.shape
```

```python
gamma = 2

for i in range(rows):
    for j in range(columns):
        img_gamma[i, j] = img[i, j] ** gamma
```

```python
plt.imshow(img_gamma, cmap="gray")
plt.show()
```
