# Logarithmic Transformation

```python
import math
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Importing the image on grayscale
img = cv.imread(r"..\Imagenes\squirrel_cls.jpg", 0)
```

```python
plt.imshow(img, cmap="gray")
plt.title("Squirrle on grayscale")
plt.axis("off")
plt.show()
```

#### For the logarthmic transformation we use the following formula  

$
s(x,y) = C \cdot \log_{10}(1 + r(x,y)) \\
\text{donde }: \\ 
r(x,y) = \text{valor de intensidad del pxel en la posicin}(x,y) \text{ de la imagen original} \\
s(x,y) = \text{valor de intensidad del pxel en la posicin}(x,y) \text{ de la imagen transformada} \\
C = \frac{255}{\log_{10}(1 + m)} \\
m = \text{pxel ms grande en la imagen}
$ 

```python
m = np.max(img)
```

```python
c = 255 / (math.log10 (1 + m))
```

```python
c
```

```python
# Using logarithms
pixel_value = 150
math.log10(pixel_value + 1)
```

```python
# Copying the image with np.zero ()
img_log = np.zeros(img.shape, dtype=np.uint8)

# Copyign the image with np.copy
img_log1 = img.copy()
```

```python
img_log.shape
```

```python
# Formula to apply the logarithmic transformation to each pixel
rows, columns = img_log.shape

for i in range(rows):
    for j in range(columns):
        img_log[i, j] = c * math.log10(img[i, j] + 1)
```

```python
plt.imshow(img_log, cmap="gray")
plt.title("Squirle with a logarithmic transformation")
plt.axis("off")
plt.show()
```
