```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

# **Threshold Transformation**
---  
## **Method 1. Iterating through all the pixels**
This transformation is useful to binarize an image

```python
img = cv.imread(r"..\Imagenes\squirrel_cls.jpg", 0)
```

```python
plt.imshow(img, "gray")
plt.show()
```

```python
img_umb = np.zeros(img.shape, dtype=np.uint8) # Is another way of copying an image
```

```python
img.shape
```

```python
row, col = img.shape
```

```python
k = 127

for i in range(row): 
    for j in range(col):
        if img[i, j] < k: # If the value of the pixel is lower than threshold it becomes 0 (black)
            img_umb[i, j] = 0
        else:
            img_umb[i, j] = 255 # If the value of the pixel is higher than threshold it becomes 255 (white)
```

```python
plt.imshow(img_umb, "gray")
plt.axis("off")
plt.show()

# The next image is the true definition of black and white
```

```python
# This is another way to threshold the image without using all the formula in the lines above
img_umb2 = cv.threshold(img, k, 255, cv.THRESH_BINARY)[1] # cv.THRESH_BINARY = 0 ## So we can subtitute that with a 0

# k = umbral
```

```python
plt.imshow(img_umb2, "gray")
plt.show()
```

```python
plt.imshow(np.hstack((img, img_umb)), "gray")
plt.title("Comparison of both images")
plt.show()
```
