### Issue with the image changing color when showing it with Matplotlib

RGB (Matplotlib)  
R - Red  
G - Green  
B - Blue  
  
BGR (OpenCV)  
B - Blue  
G - Green  
R - Red  
  
### Options to solve this issue

#### Change the order of the channels with Python

```python
plt.imshow(image[:, :, ::-1], cmap="gray")
plt.show()
```

#### Change OpenCV from openning the image from BGR to RGB

```python
image = cv.imread("..\Imagenes\gorra_diablos.png")[:, :, ::-1]

''' When you use [::-1], you are omitting the start and stop indices, which means you want to consider the entire sequence. The -1 
as the step reverses the sequence. '''

plt.imshow(image)
plt.show()
```

## Apagar ciertos canales de color

#### Red

```python
# Create a copy of the image in a new memory location. (It doesn't affect the copied image in the future)
image_r = image.copy()
```

```python
# R G B
# Shut down G and B

image_r[:, :, 1:] = 0
```

```python
image_r[:, :, 1]
```

```python
# The image is in RGB; 0 is R
plt.imshow(image_r, cmap="gray")
plt.show()
```

#### Blue

```python
# R G B
image_b = image.copy()
image_b[:, :, :2] = 0
```

```python
plt.imshow(image_b, cmap="gray")
plt.show()
```

#### Green

```python
# R G B
image_g = image.copy()
image_g[:, :, (0, 2)] = 0 # By using (0, 2) you're selecting the Red and Blue channels
```

```python
plt.imshow(image_g, cmap="gray")
plt.show()
```
