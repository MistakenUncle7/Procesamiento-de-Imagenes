# **YCbCR Space**
---

## **What is it?**
Is a **luminance-chrominance** model. It separates the image into its **luminance (Y)** component and two **chrominance (Cb, Cr)** components.

## **Components**
- Y (Luminance): Represents the **brightness** of the image.
- Cb (Chrominance Blue): Represents the **blue-difference chroma component**. It encodes how much blue there is in the color relative to the luminance (Y). Cb = B - Y (Blue minus Luminance)
- Cr (Chrominance Red): Represents the **red-difference chroma component**. It encodes how much red there is in the color relative to the luminance (Y). Cr = R - Y (Red minus Luminance)

## **How to Enhance the Contrast in the YCbCr Space**
**Steps**
1. Convert from RGB to YCbCr
2. Split the channels
3. Equalize the Y channel
4. Merge the already equalized channels --> YCbCr
5. Convert back from YCbCr to RGB
6. Display 

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
chicky = cv.imread(r"..\Images\chicky_512.png")[:, :, ::-1]
```

```python
# Step 1
chicky_ycbcr = cv.cvtColor(chicky, cv.COLOR_RGB2YCrCb)
```

```python
# Step 2
y, cb, cr = cv.split(chicky_ycbcr)
```

```python
# Step 3
y_eq = cv.equalizeHist(y)
```

```python
# Step 4
chicky_ycbcr_eq = cv.merge((y_eq, cb, cr))
```

```python
# Step 5
chicky_rgb_eq = cv.cvtColor(chicky_ycbcr_eq, cv.COLOR_YCrCb2RGB)
```

```python
# Step 6
plt.imshow(chicky_rgb_eq)
plt.title("Enhanced YCbCR space Chicky")
plt.axis("off")
plt.show()
```
