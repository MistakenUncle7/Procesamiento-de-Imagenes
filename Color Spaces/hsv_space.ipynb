# **HSV - HSB Space**
---

## **Components**
- H (Hue) --> Represents the type of color (e.g., red, blue, green)
- S (Saturation) --> Defines the intensidty or purity of the color.
-  V (Value) or B (Brightness) --> Controls the lightness or darkness of the color.

## **How to Enhance the Contrast in the HSV - HSB space**
**Steps**
1. Convert from RGB to HSV
2. Split the channels
3. Equalize the V channel (Value, holds the brightness information)
4. Convert back from HSV to RGB
5. Display

````python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
````

````python
whale = cv.imread(r"..\Images\rubberwhale1.png")[:, :, ::-1]
````

````python
PI = 3.14151926
````

````python
cv.COLOR_RGB2BGR # Is a color conversion code in OpenCV. It specifically converts an image from RGB to HSV
````

````python
# Step 1
whale_hsv = cv.cvtColor(whale, cv.COLOR_RGB2HSV) #cv.cvtColor is a function in OpenCV used to convert an image
````

#### ***cv.cvtColor*** **function**
**Components**  
- ```cv.cvtColor``` --> Is an OpenCV function that is used to convert an image from one color space to another.

**Parameters**
- ```[whale]``` --> The image you want to convert
- ```[cv.COLOR_RGB2HSV]``` --> The convertion that tells OpenCV how to transform the input image. 

### &#9888;  **You should not show HSV images in matplotlib**

````python
# You should not do the following for HSV images
plt.imshow(whale_hsv, "gray")
plt.axis("off")
plt.show()
````

### **Continuation on the steps to enhance the contrast on an HSV image**

````python
# Step 2
h, s, v = cv.split(whale_hsv)
````

````python
# Step 3
v_eq = cv.equalizeHist(v)
````

````python
whale_hsv_eq = cv.merge((h, s, v_eq))
````

````python
# This code is to show how you affected the image
plt.imshow(whale_hsv_eq)
plt.title("Effects on the Image")
plt.axis("off")
plt.show()
````

````python
# Step 4
whale_rgb_eq = cv.cvtColor(whale_hsv_eq, cv.COLOR_HSV2RGB)
````

````python
# Step 5
plt.imshow(np.hstack((whale, whale_rgb_eq)))
plt.title("Comparison Between Images")
plt.axis("off")
plt.show()
````
