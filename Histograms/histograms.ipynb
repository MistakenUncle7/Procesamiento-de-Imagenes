# **Histograms**
---
## **What is a histogram?**
Is a graphical representation that shows the distribution of data. For image processing, a histogram represents the **distribution of pixel intensities** (brightness levels) in an image. 

### **X and Y axis** 
The **x-axis** represents the intesity values -> from 0 to 255  
The **y-axis** represents the frequence of occuerence -> The number of pixels that have each intensity value

### **Why should we use a histogram?**
They provide valuable insights into the characteristics of the image, helping with various operations that improve or manipulate the image for better visual quality or for further analysis.

#### ***cv.calcHist*** **function**
**Components**  
- ```cv.calcHist``` --> OpenCV library used to calculate the histogram of an image.

**Parameters**
- ```[img]``` --> List  containing the image which you want to calculate the histogram.
- ```[0]``` --> Indicates the channel of the image for which the histogram should be calculated. If the image is grayscale you should specify ```0```
- ```None``` --> Paremter used for the mask. If you want to calculate the histogram for the entire image without any restrictions, you should set it to ```None```
- ```[256]``` --> Specifies the numbers of bins for the histogram. For an 8-bit grayscale image is appropiate because the pixel values range from ```0``` to ```256```. Each bin corresponds to one possible pixel value.
- ```(0, 256)``` --> Defines the range of pixel values for which the histogram is calculated.


````python
histrogram = cv.calcHist([img], [0], None, [256], (0, 256))
````

````python
type(histrogram)
````

````python
histrogram.shape
````

````python
# Incorrect way of printing a histogram
plt.imshow(histrogram)
plt.show()
````

````python
img.size
````

````python
# Correct way to print a histogram
plt.plot(histrogram)
plt.xlabel("Intensity")
plt.ylabel("Frequency")
plt.show()
````

# **Color Histogram**
---

````python
img = cv.imread("..\Images\WindowsLogo.jpg")
````

````python
plt.imshow(img[:, :, ::-1], "gray")
plt. title("Windows Logo")
plt.axis("off")
plt.show()
````

````python
img.shape
````

````python
histogram_b = cv.calcHist([img], [0], None, [256], (0, 256))
plt.plot(histogram_b, "b") # "b" Indicates the channel of the histogram, in this case blue.
plt.title("Blue Histogram")
plt.show()
````

````python
histogram_g = cv.calcHist([img], [0], None, [256], (0, 256))
plt.plot(histogram_g, "g") # "g" Indicates the channel of the histogram, in this case green.
plt.title("Green Histogram")
plt.show()
````

````python
histogram_r = cv.calcHist([img], [0], None, [256], (0, 256))
plt.plot(histogram_r, "r") # "r" Indicates the channel of the histogram, in this case red.
plt.title("Red Histogram")
plt.show()
````

````python
colors = ["b", "g", "r"]
````

````python
for n, i in enumerate(colors):
    temp = cv.calcHist([img], [n], None, [256], (0, 256))
    plt.plot(temp, i)


plt.title("Histogram")
plt.show()

# Explanation of the previous code on the next block of code
````
