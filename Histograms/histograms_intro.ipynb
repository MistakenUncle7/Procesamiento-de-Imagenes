```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

# Gray Scale Histograms

```python
img = cv.imread(r"..\Imagenes\blox.jpg", 0)
```

```python
plt.imshow(img, "gray")
plt.title("Blox Grayscale")
plt.show()
```

```python
histogram = cv.calcHist([img], [0], None, [256], (0, 256))
```

```python
type(histogram)
```

```python
histogram.shape
```

```python
# Incorrect way of printing a histogram
plt.imshow(histogram)
plt.show()
```

```python
# This is how a histogram should be printed
plt.plot(histogram)
plt.title("Blox Image Histogram")
plt.show()
```

## Another Gray Scale Image Histogram

```python
aero = cv.imread(r"..\Imagenes\aero3.jpg", 0)
```

```python
plt.imshow(aero, "gray")
plt.title("Aero Image Gray Scale")
plt.show()
```

```python
histogram = cv.calcHist([aero], [0], None, [256], (0, 256))
plt.title("Aero Histogram")
plt.plot(histogram)
plt.show()
```
