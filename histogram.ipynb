```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

# Grayscale histogram

####

```python
img = cv.imread(r"Imagenes\blox.jpg", 0)
```

```python
plt.imshow(img, "gray")
plt.show()
```

```python
histrogram = cv.calcHist([img], [0], None, [256], (0, 256))
```

```python
type(histrogram)
```

```python
histrogram.shape
```

```python
# Common error for a histogram
plt.imshow(histrogram)
plt.show()
```

```python
img.size
```

```python
# Correct way to print a histogram
plt.plot(histrogram)
plt.show()
```
