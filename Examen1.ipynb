**5. Utiliza la imagen anterior (la original) y visita todos los píxeles utilizando un bucle for anidado. Cuenta cuántos píxeles tienen un valor mayor a 120 en el canal de color G. Muestra el recuento.**

```python
img2 = cv.imread(r"C:\Users\Santiago Avila\Desktop\Clases\Programacion\Procesamiento de Imagenes\Images\home.jpg")[:, :, ::-1]
count = 0

rows = img2.shape[0]
columns = img2.shape[1]

for i in range(rows):
    for j in range(columns):
        if(img2[i, j, 1] > 120):
            count += 1

print(f"La cantidad de pixeles con un valor mayor a 120 en el canal verde son {count}")
```

**6. Carga una nueva imagen en escala de grises**

```python
new_img = cv.imread(r"C:\Users\Santiago Avila\Desktop\Clases\Programacion\Procesamiento de Imagenes\Images\messi5.jpg", 0)
new_img2 = new_img.copy()
new_img3 = new_img.copy()
new_img4 = new_img.copy()

plt.imshow(new_img, "gray")
plt.show()
```

Con la imagen anterior (en escala de grises), realiza las siguientes transformaciones:<br><br>
**7. Negativa. Utiliza la fórmula `s = L - 1 - r`**

```python
rows, columns = new_img.shape

for i in range(rows):
    for j in range(columns):
        new_img[i, j] = (266 - 1 - new_img[i, j])

plt.imshow(new_img, "gray")
plt.show()
```

**8. Threshold: Aplica un umbral de 127 a la imagen.**

```python
k = 127

for i in range(rows):
    for j in range(columns):
        if new_img2[i, j] < k:
            new_img2[i,j] = 0
        else:
            new_img2[i, j] = 255

plt.imshow(new_img2, "gray")
plt.show()
```

**9. Logaritmíca. A continuación las fórmulas:**<br>

$
s(x,y) = C \cdot \log_{10}(1 + r(x,y)) \\
\text{donde }: \\ 
r(x,y) = \text{valor de intensidad del píxel en la posición}(x,y) \text{ de la imagen original} \\
s(x,y) = \text{valor de intensidad del píxel en la posición}(x,y) \text{ de la imagen transformada} \\
C = \frac{255}{\log_{10}(1 + m)} \\
m = \text{píxel más grande en la imagen}
$

```python
m = np.max(new_img3)
c = (255 / math.log10(1 + m))

rows, columns = new_img3.shape

for i in range(rows):
    for j in range(columns):
        new_img3[i, j] = c * math.log10(1 + new_img3[i, j])

plt.imshow(new_img3, "gray")
plt.show()
```
